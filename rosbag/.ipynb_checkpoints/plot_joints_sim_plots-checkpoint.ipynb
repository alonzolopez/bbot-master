```python
import rosbag
import rospy
import yaml
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def bag_to_data(bagpath):
    jbagpath = bagpath # define path to bag file
    jbag = rosbag.Bag(jbagpath, 'r') # read bag file
    j_info_dict = yaml.load(jbag._get_yaml_info()) # load the yaml info file from the bag
    # for x in j1_info_dict['topics']:
    #     print(x)

    # topics of interest
    feedback = '/bbot/PositionJointInterface_trajectory_controller/follow_joint_trajectory/feedback'
    goal = '/bbot/PositionJointInterface_trajectory_controller/follow_joint_trajectory/goal'
    status = '/bbot/PositionJointInterface_trajectory_controller/follow_joint_trajectory/status'
    state = '/bbot/PositionJointInterface_trajectory_controller/state'
    state2 = '/bbot/joint_states'
    clock = '/clock'
    linkstates = '/gazebo/link_states'
    pid1 = '/bbot/gazebo_ros_control/pid_gains/bbot_joint_1/parameter_updates'
    pid2 = '/bbot/gazebo_ros_control/pid_gains/bbot_joint_2/parameter_updates'
    pid3 = '/bbot/gazebo_ros_control/pid_gains/bbot_joint_3/parameter_updates'
    pid4 = '/bbot/gazebo_ros_control/pid_gains/bbot_joint_4/parameter_updates'
    pid5 = '/bbot/gazebo_ros_control/pid_gains/bbot_joint_5/parameter_updates'
    pid6 = '/bbot/gazebo_ros_control/pid_gains/bbot_joint_6/parameter_updates'
    pid7 = '/bbot/gazebo_ros_control/pid_gains/bbot_joint_7/parameter_updates'
    paramupdates = '/gazebo/parameter_updates'
    paramdescription = '/gazebo/parameter_descriptions'

    # initialize vectors for plotting
    j_time = []
    j_desired = []
    j_actual = []
    j_error = []
    j_pid = np.zeros((7,3))

    # add data to vectors
    for topic, msg, type in jbag.read_messages():
        if topic == feedback:
            j_time.append(float(msg.feedback.header.stamp.secs) + float(msg.feedback.header.stamp.nsecs)/10.0**9)
            j_desired.append(msg.feedback.desired.positions)
            j_actual.append(msg.feedback.actual.positions)
            j_error.append(msg.feedback.error.positions)
#             print(msg)
        if topic == pid1:
            j_pid[0,0] = msg.doubles[0].value
            j_pid[0,1] = msg.doubles[1].value
            j_pid[0,2] = msg.doubles[2].value
        if topic == pid2:
            j_pid[1,0] = msg.doubles[0].value
            j_pid[1,1] = msg.doubles[1].value
            j_pid[1,2] = msg.doubles[2].value
        if topic == pid3:
            j_pid[2,0] = msg.doubles[0].value
            j_pid[2,1] = msg.doubles[1].value
            j_pid[2,2] = msg.doubles[2].value
        if topic == pid4:
            j_pid[3,0] = msg.doubles[0].value
            j_pid[3,1] = msg.doubles[1].value
            j_pid[3,2] = msg.doubles[2].value
        if topic == pid5:
            j_pid[4,0] = msg.doubles[0].value
            j_pid[4,1] = msg.doubles[1].value
            j_pid[4,2] = msg.doubles[2].value
        if topic == pid6:
            j_pid[5,0] = msg.doubles[0].value
            j_pid[5,1] = msg.doubles[1].value
            j_pid[5,2] = msg.doubles[2].value
        if topic == pid7:
            j_pid[6,0] = msg.doubles[0].value
            j_pid[6,1] = msg.doubles[1].value
            j_pid[6,2] = msg.doubles[2].value
    return j_time, j_desired, j_actual, j_error, j_pid;

```

# Joint 1 Comparison

## Data Organization

```python
j1p_bagpath = 'joint1_sim_p.bag'
j1p_time, j1p_desired, j1p_actual, j1p_error, j1p_pid = bag_to_data(j1p_bagpath)
j1p_desired1 = [pos[0]*180./math.pi for pos in j1p_desired]
j1p_actual1 = [pos[0]*180./math.pi for pos in j1p_actual]
j1p_error1 = [pos[0]*180./math.pi for pos in j1p_error]
j1p_time[:] = [t - j1p_time[0] for t in j1p_time]
```

```python
j1pd_bagpath = 'joint1_sim_pd.bag'
j1pd_time, j1pd_desired, j1pd_actual, j1pd_error, j1pd_pid = bag_to_data(j1pd_bagpath)
j1pd_desired1 = [pos[0]*180./math.pi for pos in j1pd_desired]
j1pd_actual1 = [pos[0]*180./math.pi for pos in j1pd_actual]
j1pd_error1 = [pos[0]*180./math.pi for pos in j1pd_error]
j1pd_time[:] = [t - j1pd_time[0] for t in j1pd_time]
```

## Trajectories

```python
plt.plot(j1p_time, j1p_desired1, 'k-', label="Desired")
plt.plot(j1p_time, j1p_actual1, 'r--', label="Kp = 6.64, Kd = 0.0")
plt.plot(j1pd_time, j1pd_actual1, 'g--', label="Kp = 5.0, Kd = 0.4")
plt.ylabel('Joint Angle [deg]')
plt.xlabel('Time [seconds]')
plt.title('Joint 1 Simulation: Actual vs. Desired Trajectory')
plt.legend(loc="lower right")
plt.xlim(19, 23)
plt.grid()
plt.show()
#print(j1p_pid)
```

## Error Plots

```python
plt.plot(j1p_time, j1p_error1, 'r-', label="Kp = 6.64, Kd = 0.0")
plt.plot(j1pd_time, j1pd_error1, 'g-', label="Kp = 5.0, Kd = 0.4")
plt.hlines(0.0,19.0,24.0)
plt.ylabel('Joint Angle Error [deg]')
plt.xlabel('Time [seconds]')
plt.title('Joint 1 Simulation: Trajectory Following Error')
plt.grid()
plt.legend(loc="lower right")
plt.xlim(19, 24)
plt.show()
```

# Full Trajectory

```python
motor = 1 # change this value to select which motor to plot
full_sim_bagpath_p = 'full_traj_sim_p.bag'
full_sim_time_p, full_sim_desired_p, full_sim_actual_p, full_sim_error_p, full_sim_pid_p = bag_to_data(full_sim_bagpath_p)
full_sim_desired_p = [j_d[motor-1]*180./math.pi for j_d in full_sim_desired_p]
full_sim_actual_p = [j_a[motor-1]*180./math.pi for j_a in full_sim_actual_p]
full_sim_error_p = [j_a[motor-1]*180./math.pi for j_a in full_sim_error_p]
full_sim_time_p[:] = [t - full_sim_time_p[motor-1] for t in full_sim_time_p]
full_sim_start_index_p = 0
full_sim_end_index_p = 1800

full_sim_bagpath_pd = 'full_traj_sim_pd.bag'
full_sim_time_pd, full_sim_desired_pd, full_sim_actual_pd, full_sim_error_pd, full_sim_pid_pd = bag_to_data(full_sim_bagpath_pd)
full_sim_desired_pd = [j_d[motor-1]*180./math.pi for j_d in full_sim_desired_pd]
full_sim_actual_pd = [j_a[motor-1]*180./math.pi for j_a in full_sim_actual_pd]
full_sim_error_pd = [j_a[motor-1]*180./math.pi for j_a in full_sim_error_pd]
full_sim_time_pd[:] = [t - full_sim_time_pd[motor-1] for t in full_sim_time_pd]
full_sim_start_index_pd = 0
full_sim_end_index_pd = 1800

plt.plot(full_sim_time_p[full_sim_start_index_p:full_sim_end_index_p], full_sim_desired_p[full_sim_start_index_p:full_sim_end_index_p], 'b-', label="Desired")
plt.plot(full_sim_time_pd[full_sim_start_index_pd:full_sim_end_index_pd], full_sim_actual_pd[full_sim_start_index_pd:full_sim_end_index_pd], 'r--', label="Kp = 5.0, Kd = 0.4")
plt.ylabel('Joint Angle [deg]')
plt.xlabel('Time [seconds]')
plt.title('Joint ' + str(motor) + ' Sim: Actual vs. Desired Full Trajectory')
plt.legend(loc="upper left")
print(len(full_sim_time_p))
# plt.xlim(0, 300)
# plt.ylim(-10, 40)
plt.grid()
plt.show()
```

## Error Plots

```python
plt.hlines(0.0,0.0,60.0)
plt.plot(full_sim_time_pd[full_sim_start_index_pd:full_sim_end_index_pd], full_sim_error_pd[full_sim_start_index_pd:full_sim_end_index_pd], 'b-')
plt.ylabel('Joint Angle [deg]')
plt.xlabel('Time [seconds]')
plt.title('Motor ' + str(motor) + ' Sim: Full Trajectory Error \n Kp = ' + str(full_sim_pid_pd[motor-1,0]) + ', Kd = ' + str(full_sim_pid_pd[motor-1,2]))
plt.legend(loc="upper left")
print(len(full_sim_time_p))
plt.xlim(0, 60)
# plt.ylim(-0.5, 0.5)
plt.grid()
plt.show()
```
